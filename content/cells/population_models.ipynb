```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Модель мальтуса:
Num_count = 10000

N0 = 20
a = 0.05
time = np.linspace(0, 150, Num_count)
dt = time[1]-time[0]
#print(time)

N_vals = np.zeros(np.shape(time))
N_vals[0] = N0
#print(N_vals)
for i in range(1, Num_count):
    N_vals[i] = N_vals[i-1] + N_vals[i-1]*a*dt

#print(N_vals)

fig, ax = plt.subplots()
ax.plot(time, N_vals, linewidth=2.0)
plt.show()
```

```python
# Модель Ферхюльста:
Num_count = 10000

N0 = 20
a = 0.05
time = np.linspace(0, 150, Num_count)
dt = time[1]-time[0]
#print(time)
L = 100


L_vals = np.ones(np.shape(time)) * L 
N_vals = np.zeros(np.shape(time))
N_vals[0] = N0
#print(N_vals)

# Решение методом Эйлера
for i in range(1, Num_count):
    N_vals[i] = N_vals[i-1] + N_vals[i-1]*a*(1 - N_vals[i-1]/L)*dt

#print(N_vals)

fig, ax = plt.subplots()
ax.plot(time, N_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
plt.show()
```

```python
# Модель c двумя типами:
Num_count = 10000

Cf0 = 20
a = 0.015
b = 0.1
L = 100

time = np.linspace(0, 150, Num_count)
dt = time[1]-time[0]
#print(time)


L_vals = np.ones(np.shape(time)) * L 
Cf_vals = np.zeros(np.shape(time))
Ca_vals = np.zeros(np.shape(time))
Cf_vals[0] = Cf0


# Решение методом Эйлера
for i in range(1, Num_count):
    Ca_vals[i] = Ca_vals[i-1] + Ca_vals[i-1]*b*(1 - (Ca_vals[i-1] + Cf_vals[i-1])/L)*dt + Cf_vals[i-1] * a *dt
    Cf_vals[i] = Cf_vals[i-1] - a*Cf_vals[i-1] * dt

#print(N_vals)

fig, ax = plt.subplots()
ax.plot(time, Cf_vals+Ca_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
plt.show()

fig, ax = plt.subplots()
ax.plot(time, Cf_vals+Ca_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
ax.plot(time, Ca_vals, linewidth=2.0)
ax.plot(time, Cf_vals, linewidth=2.0)
plt.show()
```

```python
# Модель c исчерпываемым ресурсом:
Num_count = 10000

Cf0 = 20
a = 0.015
b = 0.1
L = 100 # = lambda*S
mu = 0.0025

time = np.linspace(0, 200, Num_count)
dt = time[1]-time[0]
#print(time)


L_vals = np.ones(np.shape(time)) * L 
Cf_vals = np.zeros(np.shape(time))
Ca_vals = np.zeros(np.shape(time))
D_vals = np.zeros(np.shape(time))
Cf_vals[0] = Cf0


# Решение методом Эйлера
for i in range(1, Num_count):
    Ca_vals[i] = Ca_vals[i-1] + Ca_vals[i-1]*b*(1 - (Ca_vals[i-1] + Cf_vals[i-1])/(L - mu*D_vals[i-1]))*dt + Cf_vals[i-1] * a *dt
    Cf_vals[i] = Cf_vals[i-1] - a*Cf_vals[i-1] * dt
    D_vals[i] = D_vals[i-1] + (Ca_vals[i-1] + Cf_vals[i-1]) * dt

#print(N_vals)

fig, ax = plt.subplots()
ax.plot(time, Cf_vals+Ca_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
plt.show()

fig, ax = plt.subplots()
ax.plot(time, Cf_vals+Ca_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
ax.plot(time, Ca_vals, linewidth=2.0)
ax.plot(time, Cf_vals, linewidth=2.0)
plt.show()

fig, ax = plt.subplots()
ax.plot(time, D_vals, linewidth=2.0)
plt.show()
```

```python
# Модель c исчерпываемым ресурсом c учетом смертности:
Num_count = 10000

Cf0 = 20
a = 0.015
b = 0.1
d = 0.0003
L = 100 # = lambda*S
mu = 0.00

time = np.linspace(0, 200, Num_count)
dt = time[1]-time[0]
#print(time)


L_vals = np.ones(np.shape(time)) * L 
Cf_vals = np.zeros(np.shape(time))
Ca_vals = np.zeros(np.shape(time))
D_vals = np.zeros(np.shape(time))
Cf_vals[0] = Cf0


# Решение методом Эйлера
for i in range(1, Num_count):
    Ca_vals[i] = Ca_vals[i-1] + Ca_vals[i-1]*b*(1 - (Ca_vals[i-1] + Cf_vals[i-1])/(L - mu*D_vals[i-1]))*dt + Cf_vals[i-1] * a *dt - d*D_vals[i-1]*dt
    Cf_vals[i] = Cf_vals[i-1] - (a+d)*Cf_vals[i-1] * dt
    D_vals[i] = D_vals[i-1] + (Ca_vals[i-1] + Cf_vals[i-1]) * dt

#print(N_vals)

fig, ax = plt.subplots()
ax.plot(time, Cf_vals+Ca_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
plt.show()

fig, ax = plt.subplots()
ax.plot(time, Cf_vals+Ca_vals, linewidth=2.0)
ax.plot(time, L_vals,"--", linewidth=2.0)
ax.plot(time, Ca_vals, linewidth=2.0)
ax.plot(time, Cf_vals, linewidth=2.0)
plt.show()

fig, ax = plt.subplots()
ax.plot(time, D_vals, linewidth=2.0)
plt.show()
```
